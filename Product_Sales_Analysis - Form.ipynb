```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns

dataset = pd.read_csv('electronics.csv')

# list of first five rows
dataset.head(5)
```

```python
# We can see that the dataset contains 5 columns and 10000 rows.

# The columns are as follows:

# Write more description

# We can see that the columns User ID and Rating are of int64 data type, while the columns Product ID and Category are of object data type.

# We can also see that there are no null values in the dataset.

# We can also see that the column Timestamp is of int64 data type, but it is actually a timestamp.

# We can convert it to a timestamp using the following code:
dataset['timestamp'] = pd.to_datetime(dataset['timestamp'])
dataset['timestamp']
```

```python
# We can also see that the column Product ID is of object data type, but it is actually a string.

# We can convert column brand it to a string using the following code:
dataset.brand =  dataset.brand.astype("string")
```

```python
# We can also see that the column Category is of object data type, but it is actually a string.

# We can convert category it to a string using the following code:
dataset.category =  dataset.category.astype("string")
```

```python
# We can also see that the column Rating is of int64 data type, but it is actually a float.

# We can convert rating it to a float using the following code:
dataset.rating =  dataset.rating.astype("float")
```

```python
# We can also see that the column User ID is of int64 data type, but it is actually a string.

# We can convert user_id it to a string using the following code:
dataset.user_id =  dataset.user_id.astype("string")
```

```python
# We can also see that the column Product ID is of object data type, but it is actually a string.

# We can convert item_id it to a string using the following code:
dataset.item_id =  dataset.item_id.astype("string")
```

```python
# draw plot for the distribution of ratings
plt.figure(figsize=(6,5))
sns.countplot(data=dataset,x='rating' ,hue="rating",dodge=False);
```

```python
# draw visualization what was the best year of sales
sns.histplot(dataset,x='year',bins=20);
```

```python
# what brand sold the most in 2015
data_2015 = dataset[dataset.year==2015]

data_count_2015 = data_2015['brand'].value_counts()
data_top_2015=data_count_2015.head(10)
plt.figure(figsize=(12,7))
plt.bar(data_top_2015.index,data_top_2015.values)
```

```python
# Mpow sold the most followed closely with Bose while the least sold was Eldhus.
5
```

```python
# what product sold the most in 2016
data_2016 = dataset[dataset.year==2016]

data_count_2016 = data_2016['brand'].value_counts()
data_top_2016=data_count_2016.head(10)
plt.figure(figsize=(12,7))
plt.bar(data_top_2016.index,data_top_2016.values)
```

```python
# the top 3 products sold in 2016 were Bose, Logitech & TaoTronics
```

```python
# what product sold the most in 2017
data_2017 = dataset[dataset.year==2017]
data_2017 = data_2017.dropna(subset=['brand'])

data_count_2017 = data_2017['brand'].value_counts()
data_top_2017=data_count_2017.head(10)

plt.figure(figsize=(12,7))
plt.bar(data_top_2017.index,data_top_2017.values)
```

```python
# the top 3 products sold in 2017 were Etre Jeune, Panasonic,Cooper_Cases.
```

```python
# what product sold the most in 2018
data_2018 = dataset[dataset.year==2018]
data_2018= data_2018.dropna(subset=['brand'])

data_count_2018 = data_2018['brand'].value_counts()
data_top_2018=data_count_2018.head(10)

plt.figure(figsize=(12,7))
plt.bar(data_top_2018.index,data_top_2018.values)
```

```python
# the top 3 products sold in 2018 were Bose, Mpow and Logitech.
```
